```python
# Plotting graph of Product vs revenue

plt.figure(figsize=[10, 5])
plt.bar(df4["Product"],df4["Revenue"],label="Revenue", color=["blue","Red","Yellow","Green"])
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue vs Product")
plt.show()
```

Q3. Which manager earned max revenue?

```python
df.head(5)
```

```python
df["Manager"] = df["Manager"].str.strip().str.replace(r'\s+'," ",regex=True)
df["Manager"] = df["Manager"].str.title()
df.Manager.unique()
```

```python
# grouping manager with revenue
df7 = df.groupby("Manager")["Revenue"].sum().sort_values(ascending=False)
df8 = df7.reset_index()
print(df8)
```
